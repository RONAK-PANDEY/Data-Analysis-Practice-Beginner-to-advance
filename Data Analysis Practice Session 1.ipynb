```python
# Step 3: Data Cleaning

# 👉 Missing Values

# solution
data.isnull().sum()

# ans - no missing values 
```

```python
# 👉 remove the Missing Values 

# solution 
data = data.dropna() # this will remove the direct missing values (labeled as NaN, None, or NaT) from a DataFrame

# ans - removed
```

```python
# 👉 find number of the Duplicates

# solution
data.duplicated().sum()

# ans - the number of duplicates is 1 
```

```python
# remove the number of duplicates that are present 

# solution 
data = data.drop_duplicates()

# ans - removed 

print('----------------------------------------------------------------------------------------------------------')

# after removing the duplicate, you compare the new one's result vs the old one 
data.duplicated().sum()
```

```python
# Step 4: Feature Engineering 

# 1. Tip Percentage (VERY IMPORTANT FEATURE)

data["Tip_Percentage"] = (data["tip"] / data["total_bill"]) * 100

# check if that worked or not 
data.head()
```

```python
# 2. Size Category

data["size_category"] = data["size"].apply(lambda x: "small" if x <= 2 else "large")

# check if that worked or not 
data.head()
```

```python
# 2. weekend_and_working_day

# Create the correct column (overwriting any old version)
data["weekend"] = data["day"].apply(lambda x: "weekend" if x in ["Sun", "Sat"] else "working day")

# optional - Remove the redundant/incorrect column
data.drop(columns=["weekend_and_working_day"], inplace=True, errors='ignore')

# check if that worked or not 
data.head()

```

```python
# 👉 Analyze New Features

sns.boxplot(x="size_category", y="Tip_Percentage", data=data)
plt.show()

sns.countplot(x="size_category", hue="Tip_Percentage", data=data)
plt.show()

sns.lineplot(x="day", y="weekend", hue='weekend', data=data)
plt.show()
```

```python
# 5. Practice Questions (Try Yourself First!)
# 1 - Which day has the highest average tip? 
# 2 - Which time has the highest average tip?
# 3 - Do smokers tip more than non-smokers?
# 4 - Does dinner generate higher tips than lunch?
# 5 - What is the average tip percentage?
# 6 - Are larger groups tipping more (percentage-wise)?

# solution 
# first, we have to recall the things 
data.head()
```

```python
# 1 - Which day has the highest average tip?

# solution 
avg_tip_by_day = data.groupby("day")["tip"].mean()
print(avg_tip_by_day)

print("Highest avg tip day:", avg_tip_by_day.idxmax())

print("-----------------------------------------------------------------------------")
# 2 - Which time has the highest average tip?

# solution 
avg_tip_by_time = data.groupby("time")["tip"].mean()
print(avg_tip_by_time)

print("Highest avg tip day:", avg_tip_by_time.idxmax())
```

```python
# 3 - Do smokers tip more than non-smokers? 

# solution 
avg_tip_percent_by_smoker = data.groupby("smoker")["Tip_Percentage"].mean()
print(avg_tip_percent_by_smoker)

print("Highest avg tip by smoker(yes) or non smoker(no):", avg_tip_percent_by_smoker.idxmax())
```

```python
# 4 - Does dinner generate higher tips than lunch? 

# solution 
avg_tip_by_time = data.groupby("time")["tip"].mean()
print(avg_tip_by_time)

print("Highest avg tip day:", avg_tip_by_time.idxmax())

# ans - yes, the dinner generates higher tips than launch 
```

```python
# 5 - What is the average tip percentage?

# solution
data["Tip_Percentage"].mean()
```

```python
# 6 - Are larger groups tipping more (percentage-wise)?

# solution 
x = data.groupby("size_category")["Tip_Percentage"].mean()
print(x)

print("Are which groups is tipping more (percentage-wise)?", x.idxmax())

# ans - Are larger groups tipping more (percentage-wise)? no 
```
